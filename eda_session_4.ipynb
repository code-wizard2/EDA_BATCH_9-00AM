$$pie chart$$

- pie chart is a 360 degree view

- it is under matplotlib package

- it will takes keys and values

- that we will get from value counts

# step-1: get keys from continent value counts
# step-2: get from continent value counts
# step-3: plt.pie verify the arguments
# step-4: which argument takes values and keys

```python
import matplotlib.pyplot as plt
```

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.pie(values,labels=keys)
```

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.pie(values,
       explode=[0.1,0.1],
       labels=keys,
       autopct='%%0.2f%%',
       radius=1)
plt.show()
```
